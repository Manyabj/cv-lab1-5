<h1>Lab 1:Image Convolution and Edge Detection using Sobel operator</h1>
<h3>Aim:</h3>
<p>Enhancement of Features: Convolution is used to enhance important features in an image, such as edges, textures, or patterns.</p>
<p>Noise Reduction: It can help in smoothing images, reducing noise while retaining important structural information.</p>
<p>Image Transformation: Convolution allows for various transformations, such as blurring, sharpening, and detecting specific patterns.</p>
</body>
</html>

```python
!pip install opencv-python-headless
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
#load to image
image = cv2.imread(r'C:\Users\sinch\Desktop\module 2\m1.jpg')
```

```python
#display
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
plt.axis('off')  # Hide axes
plt.show()
```

```python
#apply sobel  operator for edge deection to image
if image is None:
    print("Error: Could not load image.")
else:
    # Convert the image to grayscale
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Apply the Sobel operator
    sobel_x = cv2.Sobel(gray_image, cv2.CV_64F, 1, 0, ksize=3)  # Gradient in x direction
    sobel_y = cv2.Sobel(gray_image, cv2.CV_64F, 0, 1, ksize=3)  # Gradient in y direction
    sobel = cv2.magnitude(sobel_x, sobel_y)  # Combine gradients

   
```

```python
 # Display the result
plt.imshow(sobel, cmap='gray')
plt.axis('off')
plt.title('Sobel Edge Detection')
plt.show()
```

<h1>Lab 3:Texture Analysis with Gabor Filter</h1>
<h3>Aim:</h3>
<p>Texture Feature Extraction: Gabor filters can capture different frequency components and orientations, making them effective for identifying textures in images, such as stripes, spots, and patterns.</p>
<p>Multi-Scale Analysis: By using Gabor filters at various scales and orientations, one can analyze textures at different resolutions, allowing for a comprehensive understanding of the texture structure.</p>
</body>
</html>

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread(r'C:\Users\sinch\Desktop\module 2\m1.jpg')
if image is None:
    print("Error loading image. Please check the path.")

# Convert to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Parameters for the Gabor filter
ksize = 21  # Size of the filter
sigma = 5.0  # Standard deviation of the Gaussian envelope
theta = np.pi / 4  # Orientation of the normal to the parallel stripes of the Gabor function
lamda = 10.0  # Wavelength of the cosine factor
gamma = 0.5  # Spatial aspect ratio
phi = 0  # Phase offset

# Create Gabor filter
gabor_kernel = cv2.getGaborKernel((ksize, ksize), sigma, theta, lamda, gamma, phi, ktype=cv2.CV_32F)

# Apply Gabor filter
filtered_image = cv2.filter2D(gray_image, cv2.CV_8UC3, gabor_kernel)

# Display the original and filtered images
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(gray_image, cmap='gray')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('Gabor Filtered Image')
plt.imshow(filtered_image, cmap='gray')
plt.axis('off')

plt.show()
```

<h1>Lab 4:Laplacian Edge Detection</h1>
<h3>Aim:</h3>
<p>Highlighting Rapid Intensity Changes: The Laplacian operator focuses on regions where the intensity changes rapidly, helping to isolate features such as lines, curves, and object boundaries.</p>
<p>Feature Extraction: By identifying edges, Laplacian edge detection aids in extracting features from images, which can be useful for tasks such as object recognition, image segmentation, and tracking.</p>
</body>
</html>

```python
import cv2
from matplotlib import pyplot as plt

# Load the image
image = cv2.imread(r'C:\Users\sinch\Desktop\module 2\m1.jpg', cv2.IMREAD_GRAYSCALE)

# Check if the image was loaded successfully
if image is None:
    print("Error: Could not load image.")
else:
    # Apply Laplacian edge detection
    laplacian = cv2.Laplacian(image, cv2.CV_64F)
    
    # Convert to uint8 (optional, for displaying)
    laplacian_abs = cv2.convertScaleAbs(laplacian)

    # Display original and Laplacian images
    plt.figure(figsize=(10, 5))
    
    plt.subplot(1, 2, 1)
    plt.title('Original Image')
    plt.imshow(image, cmap='gray')
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.title('Laplacian Edge Detection')
    plt.imshow(laplacian_abs, cmap='gray')
    plt.axis('off')

    plt.show()
```

<h1>Lab 5:Frequency Domain Analysis using Fourier Transforms</h1>
<h3>Aim:</h3>
<p>Understanding Periodicity: It helps identify periodic patterns within signals, making it easier to analyze phenomena that are cyclical in nature.

Time-Frequency Analysis: Tools like the Short-Time Fourier Transform (STFT) enable analysis of non-stationary signals, providing insights into how frequency content changes over time.</p>
</body>
</html>

```python


# Check if the image was loaded successfully
if image is None:
    print("Error: Could not load image.")
else:
    # Compute the 2D Fourier Transform
    f_transform = np.fft.fft2(image)
    
    # Shift the zero-frequency component to the center
    f_transform_shifted = np.fft.fftshift(f_transform)

    # Compute the magnitude spectrum
    magnitude_spectrum = np.log(np.abs(f_transform_shifted) + 1)  # +1 to avoid log(0)

    # Display the original image and the magnitude spectrum
    plt.figure(figsize=(12, 6))
    
    plt.subplot(1, 2, 1)
    plt.title('Original Image')
    plt.imshow(image, cmap='gray')
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.title('Magnitude Spectrum')
    plt.imshow(magnitude_spectrum, cmap='gray')
    plt.axis('off')

    plt.show()
```

<h1>Lab 6:Non-Maximum Suppression Result</h1>
<h3>Aim:</h3>
<p>Reduce False Edges: By suppressing non-maximum pixels, NMS minimizes the inclusion of spurious or weak edge responses that may arise from noise or other artifacts in the image.

Create Thin Edges: The result of applying NMS is typically a binary image with thin, well-defined edges, which is easier to analyze for further processing, such as object detection or recognition.

Enhance Edge Continuity: By retaining only the strongest edges, NMS contributes to a more continuous and coherent representation of object boundaries in the image.</p>
</body>
</html>

```python
# Check if the image was loaded successfully
if image is None:
    print("Error: Could not load image.")
else:
    # Apply Canny edge detector
    edges = cv2.Canny(image, 100, 200)

    # Display the original image and edges
    plt.figure(figsize=(10, 5))
    
    plt.subplot(1, 2, 1)
    plt.title('Original Image')
    plt.imshow(image, cmap='gray')
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.title('Edges with Non-Maximum Suppression')
    plt.imshow(edges, cmap='gray')
    plt.axis('off')

    plt.show()
```
